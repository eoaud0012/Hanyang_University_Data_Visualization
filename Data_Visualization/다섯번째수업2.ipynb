```python
import urllib.request
```

```python
import bs4
```

```python
import pandas as pd
```

```python
vurlobj = urllib.request.urlopen("https://en.wikipedia.org/wiki/F._Scott_Fitzgerald")
```

```python
vbsobj = bs4.BeautifulSoup(vurlobj, "html.parser")
```

```python
vbstag = vbsobj.find("div", {"class":"mw-parser-output"})
        
```

```python
vbstag
```

```python
vd = []
```

```python
for vb in vbstag.findAll("li"):
    vc = list(vb.strings)
    vd.append(vc)
```

```python
vd
```

```python
for i in range(0, len(vd), 1):
    print("il번호", i, vd[i])
```

```python
ve = []
```

```python
for i in range(20, 54, 1):
    ve.append(vd[i])
```

```python
ve
```

```python
pa = pd.DataFrame(ve)
```

```python
pa.to_csv("이대명1.csv")
```

```python
vg = []
```

```python
for vf in vbstag.findAll("ul"):
    vg.append(vf)
```

```python
vg
```

```python
for k in range(0, len(vg), 1):
    print("==================================================================")
    print("-------------ul번호--------------", k)
    print(vg[k])
```

```python
vk = []
```

```python
for i in range(2, 7, 1):
    for vi in vg[i].findAll("li"):
        vj = list(vi.strings)
        vk.append(vj)
```

```python
vk
```

```python
pb = pd.DataFrame(vk)
```

```python
pb.to_csv("이대명1두번째방법.csv")
```
