```python
import urllib.request
```

```python
import bs4
```

```python
vurlobj = urllib.request.urlopen("https://en.wikipedia.org/wiki/F._Scott_Fitzgerald")
```

```python
vbsobj = bs4.BeautifulSoup(vurlobj, "html.parser")
```

```python
vbstag = vbsobj.find("div", {"class" : "mw-parser-output"})
```

```python
vbstagg = vbsobj.find("ul")
```

```python
vbstaggg = vbsobj.findAll("li")
```

```python
vbstagggg = vbsobj.findAll("i")
```

```python
vbstagggg
```

```python
vtagatextlist = []
```

```python
for vlistfa in vbstagggg:
    vtagatext = vlistfa.get_text()
    vtagatextlist.append(vtagatext)
    print("-", vtagatext)
```

```python
vtagatextlist
```

```python
vtagatextlist = list(set(vtagatextlist))
```

```python
vtagatextlist
```

```python
import pandas as pd
```

```python
jotable = pd.DataFrame(vtagatextlist)
```
