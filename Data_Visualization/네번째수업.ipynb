```python
import bs4
```

```python
urlobj = urllib.request.urlopen("https://en.wikipedia.org/wiki/Charles_Dickens")
```

```python
bsobj = bs4.BeautifulSoup(urlobj, "html.parser")
```

```python
vlista = bsobj.findAll("i")
```

```python
vlista
```

```python
for listfor in vlista:
    print(listfor.get_text())
```

```python
distinct_list = []
```

```python
for listfor in vlista:
    distinct_list.append(listfor.get_text())
```

```python
distinct_list
```

```python
distinct_list = list(set(distinct_list))
```

```python
distinct_list
```

```python
distinct_list.remove('aa')
```
