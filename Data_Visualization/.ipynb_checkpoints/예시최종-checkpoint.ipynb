```python
import urllib.request
import bs4
urlobj = urllib.request.urlopen("https://en.wikipedia.org/wiki/F._Scott_Fitzgerald")
bsobj = bs4.BeautifulSoup(urlobj, "html.parser")
va = bsobj.find("div",{ "class" : "mw-parser-output"})
vd = []
for vb in va.findAll("li"):
    vc = list(vb.strings)
    vd.append(vc)    
```

```python
#len(vd)는 vd리스트의 구성요소의 갯수를 세는 내장함수이다. vd 리스트가 5개로 구성되어 있으면 range(0, 5, 1)이 되는 것이다.        
for i in range(0, len(vd),1):
    print("il번호", i, vd[i])
```

```python
ve=[]
for i in range(20, 54, 1):
    ve.append(vd[i])
import pandas as pd
pa = pd.DataFrame(ve)
pa.to_csv("paf.csv")
```

```python
vg = []
for vf in va.findAll("ul"):
    vg.append(vf)


for k in range(0, len(vg), 1):
    print("___________________________")
    print("ul번호",k, vg[k])
```

```python
vk=[]
for i in range(2, 7, 1):
    for vi in vg[i].findAll("li"):
        vj=list(vi.strings)
        vk.append(vj)

print(vk)    
import pandas as pd
pb = pd.DataFrame(vk)
pb.to_csv("pbf.csv")
```

```python
#이중 for문을 이해 못하는 학생을 위해  for문으로 풀어쓰기

vl=[]
for vi in vg[2].findAll("li"):
    vj=list(vi.strings)
    vl.append(vj)
for vi in vg[3].findAll("li"):
    vj=list(vi.strings)
    vl.append(vj)
for vi in vg[4].findAll("li"):
    vj=list(vi.strings)
    vl.append(vj)
for vi in vg[5].findAll("li"):
    vj=list(vi.strings)
    vl.append(vj)
for vi in vg[6].findAll("li"):
    vj=list(vi.strings)
    vl.append(vj)
print(vl)  

import pandas as pd
pc = pd.DataFrame(vl)
pc.to_csv("pcf.csv")
```
